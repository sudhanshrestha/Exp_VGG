```python
import matplotlib.pyplot as plt
import pathlib

# Define the species and their counts
species = ['belgua', 'heyna', 'leopard', 'whaleshark']

# Initialize empty lists to store counts for train, test, and valid datasets
train_counts = []
test_counts = []
valid_counts = []

# Set the paths to your data
train_data = pathlib.Path('new_dataset/train/')
test_data = pathlib.Path('new_dataset/test/')
valid_data = pathlib.Path('new_dataset/valid/')

# Count the number of images for each species in the training data
for species_name in species:
    species_folder = train_data / species_name
    num_images = len(list(species_folder.glob('*.jpg')))
    train_counts.append(num_images)

# Count the number of images for each species in the testing data
for species_name in species:
    species_folder = test_data / species_name
    num_images = len(list(species_folder.glob('*.jpg')))
    test_counts.append(num_images)

# Count the number of images for each species in the validation data
for species_name in species:
    species_folder = valid_data / species_name
    num_images = len(list(species_folder.glob('*.jpg')))
    valid_counts.append(num_images)

# Create a bar plot
plt.figure(figsize=(10, 6))
bar_width = 0.2
index = range(len(species))

plt.bar(index, train_counts, bar_width, label='Train')
plt.bar([i + bar_width for i in index], test_counts, bar_width, label='Test')
plt.bar([i + 2 * bar_width for i in index], valid_counts, bar_width, label='Valid')

plt.xlabel('Species')
plt.ylabel('Number of Images')
plt.title('Species Distribution in Train, Test, and Valid Data')
plt.xticks([i + bar_width for i in index], species, rotation=45, ha='right')
plt.legend(loc='upper right')

plt.tight_layout()

# Show the plot
plt.show()
```
